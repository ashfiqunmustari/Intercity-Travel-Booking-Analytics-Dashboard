```python
#Importing libraries
import pandas as pd
import numpy as np
from google.colab import files
```

```python
#Uploading and Loading dataset into a dataframe
uploaded = files.upload()

df = pd.read_csv("hotel_bookings.csv")
df.head()
```

```python
#Dataset overview

print("Dataset Shape:", df.shape)
print("Column Names:")
print(df.columns.tolist())

print("Missing Values:")
print(df.isnull().sum())
```

```python
#Handling Missing Values
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('Unknown')
df.drop(columns=['agent', 'company'], inplace=True)

print("Checking missing value after data handling")
print(df.isnull().sum())
```

```python
#Creating Intercity Travel Booking Features
np.random.seed(42)

operators=['Green Line', 'Shohag', 'Hanif', 'Ena', 'Saintmartin']
routes =[
        'Dhaka-Ctg',
        'Dhaka-Coxs Bazar',
        'Dhaka-Sylhet',
        'Dhaka-Khulna',
        'Dhaka-Rajshahi'
        ]

df['operator_name'] = np.random.choice(operators, size=len(df))
df['route'] = np.random.choice(routes, size=len(df))
df['seat_type'] = np.random.choice(['Economy', 'Business'],size=len(df),p=[0.8, 0.2])
df['booking_channel'] = np.random.choice(['App', 'Website'],size=len(df),p=[0.7, 0.3])
df['payment_method'] = np.random.choice(['bKash', 'Nagad', 'Card'],size=len(df))
df['number_of_seats'] = (df['adults'] + df['children']).astype(int)

df.head()
```

```python
#Renaming columns
df.rename(columns={
    'adr': 'ticket_price',
    'is_canceled': 'booking_cancelled',
    'lead_time': 'booking_lead_days',
    'is_repeated_guest': 'returning_customer'}, inplace=True)

df.head()
```

```python
#Creating Date Features
df['travel_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce')

df['booking_date'] = df['travel_date'] - pd.to_timedelta(
    df['booking_lead_days'],
    unit='D')

df['travel_month'] = df['travel_date'].dt.month_name()
df['travel_year'] = df['travel_date'].dt.year
df['travel_day_name'] = df['travel_date'].dt.day_name()

df[['booking_date', 'travel_date', 'travel_month', 'travel_day_name']].head()
```

```python
#Creating Revenue and KPI Column
df['revenue'] = df['ticket_price'] * df['number_of_seats']
df['booking_status'] = np.where(df['booking_cancelled'] == 1,'Cancelled','Completed')
df['customer_segment'] = np.where(df['returning_customer'] == 1,'Returning Customer','New Customer')
df['coupon_used'] = np.random.choice(['Yes', 'No'],size=len(df),p=[0.35, 0.65])

df.head()
```

```python
#Final columns for my analytics
final_columns = [
    'booking_date',
    'travel_date',
    'travel_year',
    'travel_month',
    'travel_day_name',
    'route',
    'operator_name',
    'seat_type',
    'booking_channel',
    'payment_method',
    'customer_segment',
    'country',
    'booking_lead_days',
    'number_of_seats',
    'ticket_price',
    'revenue',
    'booking_cancelled',
    'booking_status',
    'coupon_used',
    'market_segment',
    'customer_type',
    'previous_cancellations',
    'booking_changes',
    'deposit_type',
    'total_of_special_requests'
]

travel_df = df[final_columns].copy()

travel_df.head()
```

```python
#Initial Data Quality Check
print("Dataset Shape:", travel_df.shape)
print("\nMissing Values:")
print(travel_df.isnull().sum())
print("\nDuplicate Rows:", travel_df.duplicated().sum())

travel_df.describe()
```

```python
#Removing Invalid Records Records
print("Shape before cleaning:", travel_df.shape)

travel_df.drop_duplicates(inplace=True)
travel_df = travel_df[travel_df['ticket_price'] > 0]
travel_df = travel_df[travel_df['number_of_seats'] > 0]
travel_df['revenue'] = travel_df['ticket_price'] * travel_df['number_of_seats']

print("Shape after cleaning:", travel_df.shape)
print("Minimum Ticket Price:", travel_df['ticket_price'].min())
print("Minimum Number of Seats:", travel_df['number_of_seats'].min())
```

```python
#Creating New Business Features

travel_df['occupancy_category'] = np.where(travel_df['number_of_seats'] >= 3,'Group Booking','Individual Booking')

travel_df['loyalty_status'] = np.where(travel_df['customer_segment'] == 'Returning Customer','Loyal','New')

travel_df['trip_type'] = np.where(travel_df['route'].isin(['Dhaka-Coxs Bazar', 'Dhaka-Sylhet']),'Long Distance','Regular Route')

travel_df['revenue_category'] = pd.cut(travel_df['revenue'],bins=[0, 150, 300, 600, float('inf')],labels=['Low', 'Medium', 'High', 'Premium'])

travel_df.head()
```

```python
#Data Quality Check
travel_df.info()
travel_df.describe()

print("Missing Values:")
print(travel_df.isnull().sum())

print("Duplicate Rows:")
print(travel_df.duplicated().sum())
```

```python
#Exporting Final Dataset
travel_df.to_csv("intercity_travel_booking_analytics.csv",index=False)

print("Dataset exported successfully")
```
